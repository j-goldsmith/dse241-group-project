```python
data= pd.read_csv(r'h1b_data.csv', index_col=0)
```

```python
print(data.shape)
print(data.dtypes)
```

```python
data = data.replace(to_replace=['Y','N'], value=[1,0])
```

```python
df_dropna = data.dropna()
missing_values = data.isnull().sum().sort_values(ascending=False)
print('Total Records:',len(data))
print('Complete Records:',len(df_dropna))

print('Missing Values by Col:')
print(missing_values)
```

```python
for i in data.columns:
    try:
        data[i].hist(bins =20)
        plt.title(str(i))
        plt.show()
    except:
        continue
```

```python
date_columns = ['CASE_SUBMITTED', 'DECISION_DATE','EMPLOYMENT_START_DATE', 'EMPLOYMENT_END_DATE','ORIGINAL_CERT_DATE']
for col in date_columns:
    data[col] = pd.to_datetime(data[col])
```

```python
data['SUBMITTED_DECISION_DELTA'] = data['DECISION_DATE']-data['CASE_SUBMITTED']
data['SUBMITTED_DECISION_DELTA'].head()
```

```python
data['SUBMITTED_DECISION_DELTA'] = data['SUBMITTED_DECISION_DELTA'] / pd.Timedelta(days=1)
```
